# analise de texto 

```python
# importando as bíbliotecas
import spacy
import pandas as pd 
from collections import Counter
```

```python
#carreca os modelo em portugues 
nlp = spacy.load("pt_core_news_lg")

#carrega o texto do arquivo .txt
with open("texto.txt", "r", encoding="utf-8") as arquivo:
    texto = arquivo.read()

# processa o texto 
doc = nlp(texto)
```

```python
# listas onde informações serão armazenadas 
dados = []

# percorre todas os tokens
for token in doc:

    # ingnora pontuação e espaços
    if token.is_punct or token.is_space:
        continue

    # armazena as informações na lista
    dados.append({
        "Palavra" : token.text
        ,
        "lemma": token.lemma_,
        "classe" : token.pos_,
        "datalhe" : token.tag_
    })
```

```python
# cria o dataframe com as informações
df = pd.DataFrame(dados)
df
```

```python
df["classe"].value_counts()
```

```python
# frequencia das palavras 
frequencia = Counter(
    token.lemma_.lower()
    for token in doc
    if not token.is_punct
    and not token.is_space
    and not token.is_stop
)

df_freq = pd.DataFrame(frequencia.items(), columns=["Palavras", "Contagem"])
df_freq.sort_values("Contagem", ascending=False).head(20)
```
