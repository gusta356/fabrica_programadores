# NLP com spaCy

## doc e tokens

```python
# importando spacy
import spacy

#carregando o modelo (python -m spacy download pt_core_news_lg)
nlp = spacy.load("pt_core_news_lg")
```

```python
# processando um texto 
texto = "maria trabalha na microsoft."

doc = nlp(texto)
print(type(doc))
print(doc)
```

```python
# analizando o doc 
dados = doc.to_dict()
for k, v in dados.items():
    print(f"chave: {k:25} | valor: {type(v)}")
```

```python
# percorendo palavra por palavra 
for token in doc: 
    print(token)
```

```python
# quantidade de token
len(doc)
```

```python
# acessando um token
print(doc[0])

# texto original
print(doc.text)
```

```python
# comparando com NLTK
from nltk.tokenize import word_tokenize
import nltk

tokens = word_tokenize(texto)
print(tokens)

# spacy
doc = nlp(texto)

for token in doc:
    print(token.text)
```

## atributos dos tokens

```python
token = doc[4]
print(token)

#verifica se é composto apenas por letras (alfabéticos)
print(token.is_alpha)

# verifica se é  um stopwords
print(token.is_stop)

# verifica se é pontuação
print(token.is_punct)
```

```python
# formato da palavra

for token in doc:
    print(token, token.shape_)
```

## separação de frases 

```python
texto = """
hoje fez sol !
amanhã poderá chover.
depois voltara a fazer calor.
"""

doc = nlp(texto)
```

```python
for frase in doc.sents:
    print(frase)
```

## lematização

```python
# lematizar significa encontrar a forma "base" de uma palavra 
texto = "as crianças estava correndo felizes."
doc = nlp(texto)
for token in doc:
    print(token.text, "->", token.lemma_)
```

## clases gramaticais 

```python
texto = "o menino correu rapidamente para casa."
doc = nlp(texto)

for token in doc:
    print(token.text, "->", token.pos_)
```

## entidades 

```python
texto = "joão trabalha na microsoft em são paulo desde janeiro de 2024."
doc = nlp(texto)

for entidade in doc.ents:
    print(entidade.text, "->", entidade.label_)
```

## vetores de semelhança

```python
nlp = spacy.load("en_core_web_lg")

pairs = [
    ("car", "automobile"),
    ("dog", "cat"),
    ("brasil", "argentina"),
    ("computer", "keyboard"),
    ("banana", "airplane")
]


for  word1, word2 in pairs:
    doc1 = nlp(word1)
    doc2 = nlp(word2)

    print(f"{word1:10} x {word2:10} semelhança: {doc1.similarity(doc2)*100:.2f}%")
```
